# Functions


1. What is the diffrence between a function and a method in Python?
-  In Python, both functions and methods are blocks of reusable code designed to perform specific tasks, but their key difference lies in their association with classes and objects.
Function:
A function is a standalone block of code defined independently, not within a class.
It is called directly by its name and can operate on data passed to it as arguments.
Functions are generally used for generic operations that do not inherently belong to a particular object or class.
