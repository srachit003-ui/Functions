# 6
Explain any three AI-powered ETL (Extract, Transform, Load) tools and how they are used in analytics.
-
Three AI-powered ETL tools are Airbyte, Azure Data Factory, and AWS Glue.

▶ Airbyte - This AI capability streamlines the process of integrating data from various sources. For analytics, it means that when new data arrives, the system can automatically adjust to changes, ensuring the data remains consistent and ready for analysis without manual reconfigurations.

▶ Azure Data Factory - The AI enables the creation of resilient and efficient data pipelines, even for real-time streaming data. For analytics, this means that the data is not only moved but also continuously validated and cleaned, leading to more reliable insights and better decision-making.

▶ AWS GLUE - This tool enable the  users in finding and preparing data across their organization. For analytics, this leads to faster data preparation, as it automates many of the time-consuming steps involved in getting data ready for reporting and machine learning models.

# 8
Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI?

-
The evolution of analytics has started from manual data collection and basic reporting to advanced methods Data by Visualisation through AI Tools.

Stages are as follows:

▶ Descriptive analytics - In this Analysis, Data Analyst basically find out the Problem.

▶ Diagnostic analytics - In this Analysis, Data Analyst basically find out the root cause of an Problem

▶ Predictive analytics  - In this Analysis ,Data Analyst has to answer what might happen in the future to have some forecast and prediction so the organization can take corrective actions.

▶ Prescriptive analytics - In this Analysis , Data Analyst has to reduce the effect of future decisions in order to advise on possible outcomes before the decisions are actually made, it tries to provide how to do the actions and what are the consequences of these actions to derive a better decision.
